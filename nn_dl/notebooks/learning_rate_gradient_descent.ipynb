# üß† Understanding Learning Rate in Gradient Descent

This notebook demonstrates the impact of different learning rates on the convergence of gradient descent using a simple quadratic function: $f(x) = x^2$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## üìê The Function and Its Derivative
We define the function $f(x) = x^2$ and its derivative $f'(x) = 2x$.

```python
def f(x):
    return x ** 2


def df(x):
    return 2 * x
```

## üîÅ Gradient Descent Function
This function simulates gradient descent for a given learning rate, number of epochs, and a starting point.

```python
def gradient_descent(learning_rate, epochs, start_x):
    x = start_x
    history = [x]

    for _ in range(epochs):
        gradient = df(x)
        x -= learning_rate * gradient
        history.append(x)

    return history
```

## üß™ Experiment Setup
We'll compare how gradient descent behaves with three learning rates: `0.01`, `0.1`, and `0.9`.

```python
x = np.linspace(-10, 10, 400)
y = f(x)

start_x = 9
epochs = 20
learning_rates = [0.01, 0.1, 0.5, 0.9]

trajectories = []
for lr in learning_rates:
    history = gradient_descent(lr, epochs, start_x)
    history_y = [f(point) for point in history]
    trajectories.append((lr, history, history_y))
```

## üìä Visualization
We plot the descent paths for each learning rate on top of the function curve.

```python
plt.figure(figsize=(12, 8))
for lr, hx, hy in trajectories:
    plt.plot(hx, hy, label=f'LR: {lr}', marker='o')

plt.plot(x, y, label='f(x) = x¬≤', color='black', alpha=0.6)
plt.title('Impact of Learning Rate on Gradient Descent')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### üß† Insights:
- A small learning rate converges slowly.
- A large learning rate might overshoot the minimum or even diverge.
- A moderate learning rate finds a good balance between speed and stability.
